# ANÁLISIS DEL PROGRESO MUNDIAL DE VACUNACIÓN

##INTRODUCCIÓN
Mediante este proyecto se pretende trabajar con los datos de vacunación por pais obtenidos de la plataforma Kaggle. Se procesaran para poder determinar la cantidad de vacunas aplicadas de cada compañía, obtener la cantidad de vacunas aplicasdas en todo el mundo y calcular el promedio de vacunas aplicadas por país.


##PROCEDIMIENTO

comenzaremos por abrir el archivo de tipo .csv llamado "country_vaccinations"

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
vacunas= pd.read_csv("/content/country_vaccinations.csv")
vacunas.info()
```

Al ver la estructura y tipo de datos de cada columna se puede identificar que la columna date (fecha) se encuentra del tipo object, por lo que será necesario cambiarla a datetime64

```python
vacunas["date"]=pd.to_datetime(vacunas["date"])
vacunas.info()
```

###Cantidad de Vacunas aplicadas de cada compañia
Tomamos la columna "total_vaccinations" la cual contiene el numero total de vacunas aplicadas por dia y la columna "vaccines" que especifica el tipo de vacuna que se aplicó.

Para poder determinar la cantidad de vacunas aplicadas de cada compañia agrupamos por "vaccines", lo cual nos arroja el siguiente resultado

```python
vacunas[["total_vaccinations", "vaccines"]].groupby("vaccines").sum()
```

###Cantidad de vacunas aplicadas en todo el mundo
Para poder obtener esta información, se manda a llamar a la columna "total_vaccinations" que es la que contiene el número total de vacunas aplicadas al día, y le aplicamos la funcion sum().

```python
vacunas["total_vaccinations"].sum()
```

###Promedio de vacunas aplicadas por país

Se seleccionan las columnas "country" y "total vaccinations" y agrupamos por pais para poder obtener el promedio de vacunas que se aplicaron por pais.

```python
vacunas[["country", "total_vaccinations"]].groupby("country").mean()
```

###Cantidad de vacunas aplicadas el dia 29/01/2021 en todo el mundo

En este caso primero se mandó a llamar la columna "date" especificando que queriamos los valores que coincidieran con la fecha "2021-01-29" y lo guardamos en una variable llamada "vacunas_29_01_2021".

Posteriormente de la variable que se guardó, mandamos a llamar la columna "total_vaccinations" y sumamos los valores. Por lo tanto el resultado del total de vacunas aplicadas en todo el mundo en la fecha señalada fue de 82,952,931 vacunas.

```python
vacunas_29_01_21 = vacunas[vacunas['date'] == '2021-01-29']
total_vacunas_29_01_21 = vacunas_29_01_21['total_vaccinations'].sum()
print(f"El total de vacunas aplicadas el 29/01/21 en todo el mundo es: {total_vacunas_29_01_21}")
```

###Diferenecias de aplicacion
se crea un dataframe donde tenemos una columna llamada "diferencias" en la cual se guardan la diferencia de las cantidades de aplicacion entre "daily_vaccinations" y "daily_vaccinations_raw"

```python
conDiferencias=vacunas
conDiferencias["diferencias"]=conDiferencias["daily_vaccinations"]- conDiferencias["daily_vaccinations_raw"]
conDiferencias
```

###Periodo de tiempo de vacunanción
vamos a obtener el periodo de tiempo entre el registro con la fecha mas reciente y el registro con la fecha más antigua, para esto obtendremos el valor minimo y máximo de la columna "date" para posteriormente restarlo.

```python
fecha_minima = vacunas['date'].min()
fecha_maxima = vacunas['date'].max()
periodo_tiempo = fecha_maxima - fecha_minima
print(f"La fecha más antigua en la base de datos es: {fecha_minima}")
print(f"La fecha más reciente en la base de datos es: {fecha_maxima}")
print(f"El período de tiempo de vacunación es: {periodo_tiempo}")
```

###Cantidad de Vacunas utilizadas cada día
A continuación se crea una nueva columna donde se toma "vaccines" para poder saber que cantidad de vacunas se utilizó cada día del registro.

```python
conCantidad=vacunas
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(len)
conCantidad
```

###Registros antes del 20 de diciembre de 2020
Se selecciona del dataframe solamente los registros de vacunación aplicadas antes del 20 de diciembre de 2020 para darnos una idea de cuantas vacunas se aplicaron recién iniciadas las campañas de vacunación a nivel mundial.

```python
antes20 = vacunas[vacunas['date'] < '2020-12-20']
antes20
```

###Vacunas pfizer
Se crea un nuevo dataframe denominado Pfizer donde se guardan los datos donde la vacuna aplicada fue de la farmaceutica pfizer.

```python
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer')]
pfizer
```

###Almacenamiento de dataframes
los dataframes con los que estuvimos trabajando en este notebook se guardaran en un libro de excel denominado "resultadosReto"

```python
nombre_archivo = 'resultadosReto.xlsx'

# Crear un escritor de Excel
with pd.ExcelWriter(nombre_archivo, engine='openpyxl') as writer:
    # Almacenar cada DataFrame en una hoja separada
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

print(f'Los DataFrames se han guardado en el archivo {nombre_archivo}.')
```
